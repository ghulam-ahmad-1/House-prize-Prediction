```python
import pandas as pd 
```

```python
data =  pd.read_csv("kc_house_data.csv")
```

```python
data
```

```python
# Separating the Predicting Value : 
Y=data.iloc[:,2].values
```

```python
# Now Separating the independent Value : 
X =  data.iloc[:,3:21].values
```

```python
X
```

```python
# Data Analysis : 
from sklearn.model_selection import train_test_split
```

```python
#Visualiser le prix selon la position GPS
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize  =(12,8))
sns.scatterplot(x = 'long', y = 'lat', data=data,hue = 'price')
```

```python
X_train , X_test , Y_train ,Y_test = train_test_split(X , Y , test_size=0.3 , random_state=50)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
print(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
Random_regressor = RandomForestRegressor()
```

```python
model= Random_regressor.fit(X_train , Y_train)
```

```python
Y_predict = model.predict(X_test)
```

```python
Y_predict
```

```python
Accuracy = model.score(X_test,Y_test)
```

```python
Accuracy
```
